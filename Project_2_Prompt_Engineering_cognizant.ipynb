# **Project-1**

```python
from transformers import pipeline

# Load text-generation pipeline
generator = pipeline("text-generation", model="gpt2")

# Test Basic Prompt
basic_output = generator("Write a product description for a smartwatch.", max_length=100, num_return_sequences=1)
print("Basic Prompt Output:", basic_output[0]['generated_text'])

# Test Refined Prompt
refined_output = generator("Write a 100-word product description for a smartwatch. Highlight features like heart rate monitoring, GPS, and battery life. Use a professional tone.", max_length=100, num_return_sequences=1)
print("Refined Prompt Output:", refined_output[0]['generated_text'])

# Test Role-Based Prompt
role_output = generator("Act as a marketing expert and write a 100-word product description for a smartwatch. Highlight features like heart rate monitoring, GPS, and battery life. Use a professional yet engaging tone to appeal to fitness enthusiasts.", max_length=100, num_return_sequences=1)
print("Role-Based Prompt Output:", role_output[0]['generated_text'])
```

# **Project-2**

```python
# Test Original Prompt
original_output = generator("Explain quantum physics and include jokes.", max_length=150, num_return_sequences=1)
print("Original Prompt Output:", original_output[0]['generated_text'])

# Test Refined Prompt
refined_output = generator("Explain quantum physics in simple terms for a high school student. Include one light-hearted joke to make the topic more engaging.", max_length=150, num_return_sequences=1)
print("Refined Prompt Output:", refined_output[0]['generated_text'])
```

# **Project-3**

```python
# Generate Response
email_output = generator("Write an email inviting colleagues to a team-building event.", max_length=100, num_return_sequences=1)
print("Email Output:", email_output[0]['generated_text'])

# Evaluate Using Metrics
relevance = 4  # Does it match the task?
completeness = 3  # Does it include date, time, and location?
clarity = 5  # Is it easy to understand?
accuracy = 4  # Are details correct?
creativity = 3  # Is it engaging?
efficiency = 4  # Is it concise?

total_score = relevance + completeness + clarity + accuracy + creativity + efficiency
print("Total Score:", total_score)
```

```python
refined_email_output = generator("Write a 100-word email inviting colleagues to a team-building event on Saturday, November 18, from 10 AM to 4 PM at Central Park. Mention activities like team games and lunch. Use a friendly and enthusiastic tone.", max_length=100, num_return_sequences=1)
print("Refined Email Output:", refined_email_output[0]['generated_text'])
```
